```python
import xarray as xr

# Load the SST data
sst_dataset = xr.open_dataset('/content/AQUA_MODIS.20240801.L3m.DAY.SST.sst.4km.NRT.nc')

# Display the dataset's structure
print(sst_dataset)
```

```python
import numpy as np

# Extract SST data
sst = sst_dataset['sst'].values
lats = sst_dataset['lat'].values
lons = sst_dataset['lon'].values

# Convert the lat/lon grids into meshgrids
lon_grid, lat_grid = np.meshgrid(lons, lats)
```

```python
from scipy.spatial import cKDTree

# Define the buoy's location
buoy_lat = 34.703
buoy_lon = -72.242

# Create a KDTree for quick nearest-neighbor search
tree = cKDTree(np.c_[lon_grid.ravel(), lat_grid.ravel()])

# Find the closest point in the SST grid
distance, index = tree.query([buoy_lon, buoy_lat])
closest_lon, closest_lat = np.unravel_index(index, lon_grid.shape)

# Extract SST value for the closest grid point
sst_value = sst[closest_lat, closest_lon]

print(f"Closest SST value to buoy location ({buoy_lat}, {buoy_lon}): {sst_value}")
```

```python
import pandas as pd
import numpy as np

# Read the data
url = 'https://www.ndbc.noaa.gov/data/realtime2/41001.txt'
data = pd.read_csv(url, sep='\s+', header=None)

# Assign headers
header_row = 1  # Adjust this based on your actual data
data.columns = data.iloc[header_row]
data = data.drop([0, 1])  # Drop the header rows

# Reset index
data.reset_index(drop=True, inplace=True)

# Rename columns
data.columns = [
    '#YY', 'MM', 'DD', 'hh', 'mm', 'WDIR', 'WSPD', 'GST', 'WVHT', 'DPD', 'APD',
    'MWD', 'PRES', 'ATMP', 'WTMP', 'DEWP', 'VIS', 'PTDY', 'TIDE'
]

# Replace 'MM' with NaN
data.replace('MM', np.nan, inplace=True)

# Convert columns to numeric where possible
data = data.apply(pd.to_numeric, errors='ignore')

# Combine date and time columns into a single datetime column
data['datetime'] = pd.to_datetime(data[['#YY', 'MM', 'DD', 'hh', 'mm']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d-%H-%M')

# Set datetime as the index
data.set_index('datetime', inplace=True)

# Drop any rows that are not data (those that might have NaN or "MM" in key columns)
data.dropna(subset=['#YY', 'MM', 'DD', 'hh', 'mm'], inplace=True)

# Display the cleaned data
print(data.head())
```
